### Create single plot
  - fig, ax = plt.subplots()
    - fig = figure object
    - ax = single axes object

### Sub Plots()
  - fig, ax plots(nrow,ncol)
    - fig = figure object
    - ax = list of axes objects
    - nrow = number of rows
    - ncol = number of columns

### LEGEND
  - ax.legend(handles,labels)
    - ax = axes object
    - handles = alist of artists(line, patches) to be added to the legend
    - labels = number of columns
    - x = the horizontal(x-axis)
    - y = the vertical (y-axis)

### Line plot
  - ax.plot(x,y...)
    - x = the horizontal(x-axis)
    - y = the vertical (y-axis)


### Scatter Plot
  - ax.scatter(x,y,...)
    - x = the horizontal(x-axis)
    - y = the vertical (y-axis)
    
### Vertical bar plot
  - ax.bar(x,height..)
    - x = the horizontal(x-axis)
    - height = height of bars( y-axis)

### Horizontal bar plot
  - ax.barh(y, width...)
    -  y = the vertical(y-axis)
    - width = the width of the bars (X-axis)

### PIE chart
  - ax.pie(x,...)
    - x = the wedges size (portion)

### Box plot
  - ax.boxplot(x,...)
    - x = the input data

### Histogram
  - ax.hist(x, bins...)
    - x = the input data
    - bins = the number of bins

### Stacked area plot
  - ax.stackplot(x,y...)
    - x = the horizontal(x-axis)
    - y = the vertical (y-axis)

### Show plot
  - plt.show()

### Clear plot
- plt.cla()
- plt.clf()
- plt.close()

## Prepare Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.random import randint
```

```python
x = [1,2,3,4]
y = [20,30,60,70]
```

## Create Plot

```python
fig = plt.figure()
```

```python
#Axes
fig.add_axes()
```

```python

ax1 = fig.add_subplot(221)  #row-col-num
ax3 = fig.add_subplot(212)
```

```python
#style
plt.style.use('ggplot')
```

## Plot

```python
plt.plot(x,y)
plt.show()
```

```python
# add marker, and line style
plt.plot(x,y,color = 'y', marker = 'o', linestyle = '--')
```

```python
# add label and title
plt.plot(x,y, color = 'r', marker = '*', linestyle = '--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is line plot')
plt.show()
```

```python
# Scatter plot use random data
a = np.linspace(1,20,10)
print(a)
b = randint(1,30,10)
print(b)
```

```python
#scatter plot
plt.scatter(a,b,linewidth = 5)
plt.show()
```

```python
# Histogram plot
x = randint(1,60,40)
plt.hist(x, color = 'green')
plt.show()
```

```python
# bar plot
plt.bar(a,b, width =1.0, color ='y')
```

```python
# Horizontal bar plot
plt.barh(x, width =1.5, color ='y')
```

```python
# box plot
plt.boxplot(x, vert = True, patch_artist = True)
plt.show()
```

#CUSTOMIZE PLOT
### SUBPLOT

```python
x1 = np.array(x)
x1
```

```python
y1 = np.array(y)
y1
```

```python
y2 = y1*x1
y2
```

```python
plt.plot(x,y)
plt.plot(x, y2)
```

### xlim, ylim, xticks, and yticks

```python
fig, ax = plt.subplots(1,2, figsize = (12,4))

ax[0].plot(x1,y1,x1,y2)

ax[1].plot(x1, y2**2,'k')

ax[1].set_ylim([1000,6000])
ax[1].set_xlim([0,5])
plt.show()
```

### Add annotate

```python
fig, ax = plt.subplots(1,2, figsize = (12,4))

ax[0].annotate('Peak Value', xy=(4, 290),
                xytext=(3, 100),arrowprops=dict(facecolor='green',
                                shrink=0.05),xycoords="data",)
ax[0].plot(x1,y1,x1,y2)

ax[1].plot(x1, y1**2, 'k')


```

# PIE chart & savefig

```python
# PIE chart & savefig & add legend
labels = ['Ayam', 'Ikan', 'Kura-kura', 'Harimau']
sizes = [20, 43, 56, 75]

fig, ax = plt.subplots(figsize = (4, 4), dpi = 100)
explode = (0.1, 0, 0, 0)
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90, explode = explode)
ax.legend()
plt.savefig
plt.show()
```

## Line Plot on a polar axis & Show plot

```python
#Show plot
r = np.arange(0, 7, 0.05)
theta = 2*np.pi*r
print(r)

plt.figure(dpi = 100)
ax = plt.subplot(111, projection = 'polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0, 0.5, 1, 1.5])
ax.set_rlabel_position(-30)
ax.grid(True)
plt.show()
```
